```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
pokemon = pd.read_csv("../data/chipotle_stores.csv")
X = pokemon[["latitude", "longitude"]]
```

```python
model = DBSCAN(eps=0.05, min_samples=1).fit(X)
model = model.fit(X)
pred = model.fit_predict(X)
```

```python
plt.title("DBSCAN")
plt.scatter(pokemon.longitude, pokemon.latitude, c=pred)
plt.show()
```

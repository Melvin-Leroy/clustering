```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram
from sklearn.cluster import AgglomerativeClustering
```

```python
pokemon = pd.read_csv("../data/chipotle_stores.csv")
X = pokemon[["latitude", "longitude"]]
```

```python
model = AgglomerativeClustering(n_clusters=35, affinity="euclidean", linkage="complete")
model = model.fit(X)
pred = model.fit_predict(X)
```

```python
pred
```

```python
model.labels_
```

```python
from scipy.spatial.distance import pdist

y = pdist(X)
y
```

```python
plt.title("Agglomerative Clustering")

plt.scatter(pokemon.longitude, pokemon.latitude, c=pred)
plt.legend()

plt.show()
```
